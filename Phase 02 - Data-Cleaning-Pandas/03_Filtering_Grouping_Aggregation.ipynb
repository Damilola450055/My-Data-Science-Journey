##### Filtering and sorting

```python
Data = {
        'id': ['A1', 'A2', 'A3', 'A4', 'A5'],
        'project_id_initial': ['DX13', 'DX11','DX29', 'DX30', 'DX28'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'role': ['Python Developer', 'Python Developer', 'PHP Developer', 'C++ Developer', 'Data '],
        'experience': [3,4,4,3,2],
        'qualification': ['Graduate', 'Graduate', 'Post Graduate', 'Graduate', 'Post Graduate']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(Data)
df
```

```python
Role_df = df[df['role'] == 'Python Developer']
Role_df
```

```python
Experience_df = df[df['experience'] > 3]
Experience_df
```

##### Slicing

```python
# selecting sub set of a row or column
# loc slicing
df
```

```python
df.loc[:, ['id', 'role','experience']]
```

```python
df.loc[1:3, ['id', 'role','experience']] # label indexing
```

```python
# iloc slicing
df
```

```python
df.iloc[:, [0, 3, 4]] # interger indexing 
```

#### Grouping data and aggregate operations 

```python
df.groupby('role')['experience'].sum()
```

```python
df.groupby(['id', 'role'])['experience'].sum()
```

```python
df.groupby(['id', 'role'])['experience'].agg(['sum', 'mean', 'min', 'max'])
```
