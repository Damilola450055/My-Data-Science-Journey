# Data Visualization using Python Libraries -- Pandas, Matplotlib and Seaborn


![](https://i.imgur.com/9i806Rh.png)


Data visualization is the graphic representation of data. It involves producing images that communicate relationships among the represented data to viewers. Visualizing data is an essential part of data analysis and machine learning.

| **Chart Type**       | **Best For**                                      | **Example Use Case**                          | **Python Code Example**                     | **When to Avoid** | **Best Practices** |
|----------------------|--------------------------------------------------|-----------------------------------------------|---------------------------------------------|-------------------|--------------------|
| **Bar Chart**        | Comparing categorical data.                      | Sales by product category.                    | `sns.barplot(x='category', y='sales')`      | When you have too many categories (>15) | Order bars by value, use consistent colors |
| **Histogram**        | Showing distribution of numerical data.          | Age distribution of survey respondents.       | `plt.hist(df['age'], bins=20)`              | For small datasets (<30 points) | Choose appropriate bin size, remove outliers first |
| **Box Plot**         | Displaying spread & outliers in numerical data.  | Comparing salaries across departments.        | `sns.boxplot(x='dept', y='salary')`         | When you need exact values | Add swarmplot for small datasets |
| **Scatter Plot**     | Revealing relationships between 2 numerical vars.| Height vs. weight correlation.                | `plt.scatter(df['height'], df['weight'])`   | With >10,000 points (use hexbin) | Add regression line, use alpha for dense data |
| **Line Chart**       | Tracking trends over time (time-series data).    | Monthly website traffic.                      | `plt.plot(df['month'], df['traffic'])`      | With irregular time intervals | Label x-axis clearly, limit to <5 lines |
| **Pie Chart**        | Showing proportions of a whole (use sparingly!). | Market share by company.                      | `df.plot.pie(y='market_share')`             | When you have >5 categories or similar values | Order slices by size, limit to 3-5 categories |
| **Heatmap**          | Visualizing correlation matrices or density.     | Feature correlations in a dataset.            | `sns.heatmap(df.corr())`                    | With too many variables (>15) | Use diverging colormap, annotate values |
| **Violin Plot**      | Combining distribution and density.              | Distribution of exam scores by gender.        | `sns.violinplot(x='gender', y='score')`     | For very small datasets | Combine with boxplot for quartiles |
| **Pair Plot**        | Exploring multiple pairwise relationships.       | Quick EDA of numerical features.              | `sns.pairplot(df, hue='target')`            | With >10 variables | Use for initial exploration only |
| **Count Plot**       | Showing frequency of categorical observations.   | Number of customers by region.               | `sns.countplot(x='region')`                 | When categories have similar counts | Order by frequency, use horizontal for long labels |
| **Area Chart**       | Tracking cumulative totals over time.            | Quarterly revenue accumulation.               | `df.plot.area(x='quarter', y='revenue')`    | With many overlapping categories | Use stacked carefully, limit to 4-5 categories |
| **Bubble Chart**     | Comparing 3+ variables (x, y, and bubble size).  | GDP vs. life expectancy with population size. | `plt.scatter(x, y, s=size*10, alpha=0.5)`   | When size differences are subtle | Scale bubbles appropriately, add legend |
| **Swarm Plot**       | Showing individual observations with distribution| Customer age by subscription type.           | `sns.swarmplot(x='type', y='age')`         | Large datasets (>1000 points) | Combine with violin/box plot |
| **Hexbin Plot**      | Visualizing dense scatter plots.                | Geographic coordinate density.               | `plt.hexbin(x,y, gridsize=20)`             | Sparse data | Adjust gridsize for clarity |
| **Radar Chart**      | Comparing multivariate data points.             | Player skill attributes.                     | `plotly.express.line_polar()`              | When variables aren't comparable | Normalize scales, limit to <8 variables |
| **Treemap**         | Showing hierarchical part-to-whole relationships| Budget allocation by department.             | `squarify.plot(sizes=values)`              | Deep hierarchies (>3 levels) | Use color for additional dimension |
| **Sankey Diagram**  | Visualizing flow or transitions.                | Customer journey between stages.             | `plotly.express.sankey()`                  | Complex flows (>10 nodes) | Highlight key paths |
| **Gantt Chart**     | Project timelines and schedules.                | Project task durations and dependencies.     | `plotly.express.timeline()`                | When tasks aren't time-based | Use colors for categories |
| **Word Cloud**      | Visualizing text data frequency.               | Most common words in customer reviews.       | `WordCloud().generate(text)`               | For precise comparisons | Remove stopwords first |
| **Calendar Heatmap**| Temporal patterns in data.                     | Daily website visits over a year.            | `calmap.calendarplot()`                    | Irregular time periods | Use intuitive colormap |

```python
#import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Creating a Simple Plot
> **Line Plot**

```python
apples_yield = [0.895, 0.91, 0.919, 0.926, 0.931, 0.928]
plt.plot(apples_yield);
```

```python
apples_yield = [0.895, 0.91, 0.919, 0.926, 0.931, 0.928]
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, apples_yield);
```

```python
apples_yield = [0.895, 0.91, 0.919, 0.926, 0.931, 0.928]
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, apples_yield)
plt.xlabel('Years')
plt.ylabel('Apple Yield')
plt.title('Yearly yield rate of Apple')
plt.show()
```

### Plotting multiple lines

```python
years = range(2010, 2016)
apples_yield = [0.895, 0.91, 0.919, 0.926, 0.931, 0.928]
oranges_yield = [0.99, 0.978, 0.985, 0.908, 0.889, 0.878]
```

```python
plt.plot(years, apples_yield, marker='o')
plt.plot(years, oranges_yield, marker='*')
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly yield rate of Apple and Oranges')
plt.legend(['apple', 'orange'])
plt.show()
```

### Styling lines and marker

```python
plt.plot(years, apples_yield, marker='o', c='red', ls='--', lw=0.5, ms=12, mec='navy', mew=6)
plt.plot(years, oranges_yield, marker='*', c='green', ls='-', lw=4, ms=4, mec='blue', mew=4)
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly yield rate of Apple and Oranges')
plt.legend(['apple', 'orange'])
plt.show()
```

FMT= '[marker][line][color]'

```python
plt.plot(years, apples_yield, '*-b', lw=1)
plt.plot(years, oranges_yield, 'o--g')
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly yield rate of Apple and Oranges')
plt.legend(['apple', 'orange'])
plt.show()
```

#### Changing the Figure Size

```python
plt.figure(figsize=(5, 3))

plt.plot(years, apples_yield, '*-b')
plt.plot(years, oranges_yield, 'o--g')
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly yield rate of Apple and Oranges')
plt.legend(['apple', 'orange'])
plt.show()
```

```python
plt.figure(figsize=(8, 3))

plt.bar(years, apples_yield)
plt.bar(years, oranges_yield, bottom=apples_yield)
```

```python
# sum of tip per day
tips_sum = df.groupby('day')['tip'].sum()
```

```python
tips_sum
```

```python
plt.figure(figsize=(5, 3))

tips_sum.plot(kind='bar')
plt.title('Total tips per day', fontweight='bold')
plt.ylabel('Total tips')
plt.xlabel('Day of week')
plt.xticks(rotation=0)
plt.show()
```

```python
# Scatter plot
sns.set_style('whitegrid')
plt.figure(figsize=(6, 3))

plt.scatter(data=df, x='tip', y='total_bill', color='red', alpha=0.7)
plt.xlabel('Tip')
plt.ylabel('Total bills')
plt.grid(c='gray', ls='--', alpha=0.3)
plt.show()
```
