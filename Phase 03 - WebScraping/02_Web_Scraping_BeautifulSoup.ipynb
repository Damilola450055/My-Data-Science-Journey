```python
url='https://www.who.int/health-topics/#A'

response = requests.get(url)
response.status_code
soup = BeautifulSoup(response.text, 'html.parser')

topics = []


for tag in soup.find_all('a', class_="link-container table"):
    topic_name = tag.find('p', class_="heading text-underline").get_text()
    topic_link = tag['href']
    topics.append({
        'name': topic_name,
        'link': topic_link
    }) 

# collect data for each topic
Data = []

for topic in topics:
    print(f'Scraping content for: {topic["name"]}')  
    link = topic['link']
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # locate tags
    contents =soup.find_all('div', class_="sf_colsOut tabContent")

    #create heading
    Heading = {
        'Overview' : 'N/A',
        'Impact' : 'N/A',
        'Who response' : 'N/A'

    }
    if len(contents)>0:
        Heading['Overview'] = contents[0].get_text(strip=True)
    if len(contents)>1:
        Heading['Impact'] = contents[1].get_text(strip=True)
    if len(contents)>2:
        Heading['Who response'] = contents[2].get_text(strip=True)


    Data.append({
    'Health topic': topic['name'],
    'Links': topic['link'],
    'Overviews': Heading['Overview'],
    'Impacts': Heading['Impact'],
    'Who responses': Heading['Who response']
    })
```

```python
df = pd.DataFrame(Data)
df
```
