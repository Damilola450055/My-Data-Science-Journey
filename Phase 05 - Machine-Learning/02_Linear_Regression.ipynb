##### formular for linear line y=mx+c5

```python
# we split into features and target variable
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
!pip install scikit-learn
```

```python
#fiiting in the best line
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
%%time
model.fit(x, y)
```

##### Prediction

```python
prection = model.predict(x)
prection
```

```python
prection[1]
```

```python
y[1]
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x='size', y='price', data=df)
plt.plot(x, prection, color='red')
plt.xlabel('size')
plt.ylabel('price')
plt.title('the relationsip between the size and the price')
plt.show()
```

```python
c = model.intercept_ 
c
```

```python
m = model.coef_
m
```

```python
x = 643.09 # size
# y-mx+c
Predicted_price = m*x+c
Predicted_price
```
