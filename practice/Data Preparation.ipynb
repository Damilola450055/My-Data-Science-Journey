```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_temp = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'New York': [32, 35, 45, 58, 68, 77, 82, 80, 73, 62, 48, 38],
    'Los Angeles': [68, 70, 72, 75, 78, 82, 85, 86, 84, 79, 72, 68],
    'Chicago': [22, 26, 38, 51, 62, 72, 77, 75, 67, 55, 40, 28]
}

temp = pd.DataFrame(data_temp)
print(temp)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(temp.iloc[:, 0], temp.iloc[:, 1], 'o--r');
```

```python
plt.figure(figsize=(12, 6))
plt.plot(temp.iloc[:, 0], temp.iloc[:, 2], 's-.g')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(temp.iloc[:, 0], temp.iloc[:, 3], 'D:b')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=temp, x='Month', y='New York', label='New York')
sns.lineplot(data=temp, x='Month', y='Los Angeles', label='Los Angeles')
sns.lineplot(data=temp, x='Month', y='Chicago', label='Chicago')

plt.title("Monthly Temperatures in U.S. Cities")
plt.xlabel("Month")
plt.ylabel("Temperature (°F)")
```
